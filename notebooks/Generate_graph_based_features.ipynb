This notebook generates part of the graph-based features. Other part is generated by files in the src directory.

```python
import pandas as pd
import networkx as nx
import numpy as np
import pickle
```

```python
# loading the graph as a netoworkx object
with open(r'..\data\interim\data.pickle','rb') as file:
        data = pickle.load(file)
```

```python
# setting up a dataframe, adding node outdegree to it, and saving it in a csv file
df=pd.DataFrame(list(data.out_degree))
df.rename(columns = {0:'paper_id', 1:'outdegree'}, inplace = True)
```

```python
# calculating cumulative outdegrees of papers' successors
def outdegrees_sum_successors(current_paper) -> int:
    cumulative_outdegree_of_cited_papers = 0  
    cited_by_current_paper = data[current_paper]
    for cited_paper in cited_by_current_paper:
        cumulative_outdegree_of_cited_papers += data.out_degree(cited_paper)
    print('paper ',current_paper,' processed, cumulative outdegree of cited papers is ',cumulative_outdegree_of_cited_papers)
    return cumulative_outdegree_of_cited_papers
        
df["sum_outdegrees_successors"] = df["paper_id"].apply(lambda current_paper: outdegrees_sum_successors(current_paper))
```

```python
# calculating averages of papers' successors' outdegrees
df['avg_outdegree_successors']=df['sum_outdegrees_successors']/df['outdegree']
df.loc[df['avg_outdegree_successors'].isnull(), 'avg_outdegree_successors'] = 0
```
