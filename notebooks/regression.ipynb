```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import datetime
```

```python
df=pd.read_csv(r'../data/processed/all_features_and_target_variabes.csv')
df=df.drop(columns=['Unnamed: 0'])
#df=df.dropna()
try:
    df['first_submission_date'] = df["first_submission_date"].apply(lambda x: datetime.datetime.strptime(x, '%Y-%m-%d'))
except TypeError:
    pass
#datetime_object = datetime.strptime('Jun 1 2005  1:33PM', '%b %d %Y %I:%M%p')
df
```

```python
features=df[['indegrees_of_cited_papers_sum',
             'indegrees_of_cited_papers_avg',
             'recency_of_cited_papers_avg',
             'max_time_difference_bw_cited_papers',
             'outdegree',
             'outdegrees_of_cited_papers_sum',
             'outdegrees_of_cited_papers_avg',
             'is_revised',
             'times_revised',
             'num_of_pages',
             #'format',
             'journal_counts'
            ]]
```

```python
features.isna().sum()
```

```python
features['recency_of_cited_papers_avg'] = features['recency_of_cited_papers_avg'].fillna(0)
features['max_time_difference_bw_cited_papers'] = features['max_time_difference_bw_cited_papers'].fillna(0)
features['num_of_pages'] = features['num_of_pages'].fillna(0)
features['journal_counts'] = features['journal_counts'].fillna(0)
features.isna().sum()
```

```python
target=df[['first_365_days']]
target['first_365_days'] = target['first_365_days'].fillna(0)
target.isna().sum()
```

```python
features_train = features[[]]
    features_train, features_test, target_train, target_test = train_test_split(
     features, target, test_size=0.33, random_state=42)
```

```python
# create and fit a linear regression
estimator = LinearRegression()
estimator.fit(features_train, target_train)
prediction = estimator.predict(features_test)
```

```python
# evaluating
from sklearn.metrics import mean_squared_error, r2_score
from math import sqrt


mse_1 = mean_squared_error(target_test, prediction)
r2_1 = r2_score(target_test, prediction)
print("MSE:", mse_1)
print("RMSE:", sqrt(mse_1))
print("R^2:", r2_1)
print()
```
