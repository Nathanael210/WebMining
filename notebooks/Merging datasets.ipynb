```python
import pandas as pd
```

```python
graph_features_1=pd.read_csv(r'../data/processed/metadata_and_graphfeat_part1.csv')
graph_features_1=graph_features_1.drop(columns=['Unnamed: 0', 'Unnamed: 0.1','submitter','submission_date','title','authors','comments','report_no','journal_ref','abstract','outdegree'])
graph_features_1=graph_features_1.rename(columns={
    "importance_of_cited_papers(sum)": "indegrees_of_cited_papers_sum",
    "importance_of_cited_papers(avg)": "indegrees_of_cited_papers_avg",
    "recency(avg)": "recency_of_cited_papers_avg",
    "max_time_diff_cit": "max_time_difference_bw_cited_papers"
})
graph_features_1
```

```python
graph_features_2=pd.read_csv(r'../data/processed/graph_features_2.csv')
graph_features_2=graph_features_2.drop(columns=['Unnamed: 0'])
graph_features_2=graph_features_2.rename(columns={
    "sum_outdegrees_successors": "outdegrees_of_cited_papers_sum",
    "avg_outdegree_successors": "outdegrees_of_cited_papers_avg"
})
graph_features_2
```

```python
metadata_features=pd.read_csv(r'../data/processed/metadata_features.csv')
metadata_features=metadata_features.drop(columns=['Unnamed: 0'])
metadata_features
```

```python
target_variables=pd.read_csv(r'../data/processed/papers_label.csv')
target_variables=target_variables.rename(columns={"Unnamed: 0": "paper_id"})
target_variables
```

```python
all_features=graph_features_1.merge(graph_features_2, on='paper_id', how='left')
all_features=all_features.merge(metadata_features, on='paper_id', how='left')
all_features.to_csv(r'../data/processed/all_features.csv')
all_features=all_features.merge(target_variables, on='paper_id', how='left')
all_features.to_csv(r'../data/processed/all_features_and_target_variabes.csv')
```
